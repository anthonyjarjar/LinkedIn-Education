Assume the type of avocado and price is random chance

Test Statistc: (Price of Organic Avocado) - (Price of Conventional Avocado)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np
```

```python
avocadoData = pd.read_csv('avocado.csv')
avocadoData
```

```python
conventional = avocadoData.loc[avocadoData['type'] == 'conventional', :]
organic = avocadoData.loc[avocadoData['type'] == 'organic']
```

```python
observedTestStat = np.mean(conventional['AveragePrice'] - np.mean(organic['AveragePrice']))
observedTestStat
```

```python
# randomly premute the datasample using .sample
avocadoData['AveragePrice'].sample(frac = 1)
```

```python
# new set of indicies with old set of indicies
avocadoData['AveragePrice'].sample(frac = 1).reset_index()
```

```python
# drops old indicies
avocadoData['AveragePrice'].sample(frac = 1).reset_index(drop = True)
```

```python
# permutes the data more than once
def perm(data):
    return data.sample(frac = 1).reset_index(drop = True)
```

```python
simTestStat = np.array([])

reps = 10000

for i in range(reps):
    permInfo = perm(avocadoData['AveragePrice'])
    
    df = pd.DataFrame({'Permuted Duration' : permInfo, 'Type' : avocadoData['type']})
    
    conv = df.loc[df['Type'] == 'conventional', 'Permuted Duration']
    org = df.loc[df['Type'] == 'organic', 'Permuted Duration']
    
    stat = np.mean(conv) - np.mean(org)
    
    simTestStat = np.append(simTestStat, stat)
    
# randomizing the sample data and assigning types (Organic or Conventional)
# to the random data to see if there truly is a cooreltation with
# the data or not 
```

```python
df
```

```python
simTestStat
```

```python
pValue = np.count_nonzero(simTestStat <= observedTestStat) / reps
pValue
```

The pValue = 0, meaning,  approximately about zero percent of the 10,000 permuted samples of the data resulted in a test statstic with the same value of the observed test stastic or lower 

So basically, randomizing the the data does not give the same result as the data, meaning their is a coorelation  

We can now reject the null hypothesis
