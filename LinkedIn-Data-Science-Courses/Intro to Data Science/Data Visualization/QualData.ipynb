```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
listing = pd.read_csv('Airbnb_NYC_2019.csv')
listing
```

```python
sn.countplot(x = 'neighbourhood_group', data = listing)
plt.show()
```

```python
sn.barplot(x = 'neighbourhood_group', y = 'price', data = listing) # black bar is the confidence interval
plt.show()
```

```python
sn.barplot(x = 'neighbourhood_group', y = 'price', data = listing, ci = False)
plt.show()
```
