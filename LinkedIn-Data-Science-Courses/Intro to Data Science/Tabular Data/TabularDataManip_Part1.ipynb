```python
import pandas as pd
```

```python
usBabies = pd.read_csv("us_baby_names.csv")
```

```python
usBabies
```

```python
# what were the five most popular baby names in 2014
# first slice out rows for 2014
# sort rows in decending order by count 
# get the first five rows of the data frame
```

```python
usBabies['Year']
```

```python
 usBabies['Year'] == 2014 # returns boolean value, true if year = 2014
```

```python
usBabies2014 = usBabies.loc[usBabies['Year'] == 2014, :] # acess all the rows where year is 2014
usBabies2014
```

```python
sorted2014 = usBabies2014.sort_values('Count', ascending = False)
sorted2014
```

```python
sorted2014.iloc[0:5] # completed the problem!
```

```python
# What were the most popular baby names in california during 2014
# We will use the same method as before
```

```python
stateBabies = pd.read_csv('state_baby_names.csv')
```

```python
stateBabies
```

```python
stateYear = stateBabies.loc[stateBabies['Year'] == 2014, :]
stateYear
```

```python
stateFiltered = stateYear.loc[stateYear['State'] == 'CA', :]
stateFiltered
```

```python
sortedCA2014 = stateFiltered.sort_values('Count', ascending = False)
sortedCA2014
```

```python
sortedCA2014[0:5]
```
