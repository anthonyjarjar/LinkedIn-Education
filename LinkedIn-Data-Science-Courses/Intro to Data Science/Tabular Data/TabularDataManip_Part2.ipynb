```python
import pandas as pd
```

```python
# In california, what were the most popular female and male baby names for each year
```

```python
stateBabies = pd.read_csv('state_baby_names.csv')
```

```python
caBabies = stateBabies.loc[stateBabies['State'] == 'CA', :]
caBabies
```

```python
def popular(s):
    """Receives s, a pandas series containing baby names in order of highest count to lowest count.
    Returns the most popular name s."""
    return s.iloc[0]
```

```python
caGrouped = caBabies.sort_values('Count', ascending = False).groupby(['Year', 'Gender']).agg(popular)

# sorts entire date frame in decending order, to get most popular names 

# groups data frame into years and genders, i.e. (Data frame for the year 1910, contaning male names)

# .agg(popular) indexes the first row of the grouped data frame and aggregated them into one dataframe

caGrouped
```
